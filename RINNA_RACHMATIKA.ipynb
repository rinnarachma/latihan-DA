# **MENGHITUNG STANDAR DEVIASI DARI DATAFRAME**

```python
print ("Rata-rata CreditScore")
print (churn['CreditScore'].mean())

print ("Rata-rata Age")
print (churn['Age'].mean())

print ("Rata-rata Tenure")
print (churn['Tenure'].mean())

print ("Rata-rata Balance")
print (churn['Balance'].mean())

print ("Rata-rata EstimatedSalary")
print (churn['EstimatedSalary'].mean())
```

```python
print ("Median CreditScore")
print (churn['CreditScore'].median())

print ("Median Age")
print (churn['Age'].median())

print ("Median Tenure")
print (churn['Tenure'].median())

print ("Median Balance")
print (churn['Balance'].median())

print ("Median EstimatedSalary")
print (churn['EstimatedSalary'].median())
```

```python
# Import Libray
import matplotlib.pyplot as plt
import numpy as np

# Membuat data dummy
mu, sigma = 0.5, 0.1
s = np.random.normal(mu, sigma, 1000)

# Membuat histogram
count, bins, ignored = plt.hist(s, 20)

# Plot grafik distribusi
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=3, color='y')
plt.show()
```
