# **MENGHITUNG STANDAR DEVIASI DARI DATAFRAME**

Terdapat beberapa rata-rata dari data yang telah dihitung berdasarkan CreditScore:

2.   Rata-rata Usia (Age): 38.9218 : Ini adalah rata-rata usia dari individu. Informasi ini bisa memberi pemahaman tentang distribusi usia pelanggan atau individu dalam dataset.

3.   Rata-rata Tenure: 5.0128 : Rata-rata masa kerja (tenure) menunjukkan berapa lama rata-rata pelanggan atau individu telah menjadi pelanggan atau bergabung dengan suatu layanan atau perusahaan. Ini bisa memberi wawasan tentang loyalitas pelanggan atau tingkat retensi.

4.   Rata-rata Balance: 76485.889288 : Rata-rata saldo menunjukkan jumlah rata-rata uang yang tersisa dalam akun atau portofolio keuangan dari individu. Ini bisa memberikan gambaran tentang tingkat kekayaan atau kesejahteraan finansial dari sampel tersebut

5.   Rata-rata EstimatedSalary: 100090.239881 : Ini menunjukkan perkiraan gaji rata-rata dari individ. Informasi ini bisa memberi pemahaman tentang tingkat pendapatan dari orang-orang.

```python
# Import Libray
import matplotlib.pyplot as plt
import numpy as np

# Membuat data dummy
mu, sigma = 0.5, 0.1
s = np.random.normal(mu, sigma, 1000)

# Membuat histogram
count, bins, ignored = plt.hist(s, 20)

# Plot grafik distribusi
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=3, color='y')
plt.show()
```

Histogram yang menunjukkan distribusi frekuensi dari data acak yang telah dibuat, bersamaan dengan plot garis yang mewakili distribusi normal yang sebenarnya dapat memungkinkan untuk membandingkan distribusi data acak dengan distribusi normal yang diharapkan dari dataset.

# **KORELASI**

```python
import pandas as pd

# Ambil subset data yang ingin Anda analisis
subset_data = churn[['Exited', 'CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']]

# Hitung korelasi Pearson
pearson_corr = subset_data.corr()

# Hitung korelasi Spearman
spearman_corr = subset_data.corr(method='spearman')

print("Korelasi Pearson:")
print(pearson_corr)

print("\nKorelasi Spearman:")
print(spearman_corr)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Hitung korelasi
corr = subset_data.corr()

# Plot heatmap korelasi dengan Matplotlib
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(np.arange(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(np.arange(len(corr.index)), corr.index)
plt.title('Heatmap Korelasi')
plt.show()
```

Setiap sel pada heatmap direpresentasikan dengan warna yang mencerminkan nilai korelasi antara dua variabel. Biasanya, warna yang lebih gelap menunjukkan korelasi yang lebih tinggi, sementara warna yang lebih terang menunjukkan korelasi yang lebih rendah atau tidak ada korelasi.

```python
import seaborn as sns

# Plot heatmap korelasi dengan Seaborn
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi')
plt.show()
```

Koefesien
Korelasi menunjukkan kekuatan (strength) hubungan linear dan arah hubungan. Jika koefesien korelasi positif, maka kedua variabel mempunyai
hubungan searah. Artinya jika nilai variabel X tinggi, maka nilai variabel
Y akan tinggi pula. Sebaliknya, jika koefesien korelasi negatif, maka kedua variabel
mempunyai hubungan terbalik. Artinya jika nilai variabel X tinggi, maka
nilai variabel Y akan menjadi rendah (dan sebaliknya). Warna semakin gelap artinya semakin berkorelasi. Tapi interpretasinya mengikuti ini Bu:
0 : Tidak ada korelasi antara dua variabel
>0 0,25 : Korelasi sangat lemah
>0,25 0,5 : Korelasi cukup
>0,5 0,75 : Korelasi kuat
>0,75 0,99 : Korelasi sangat kuat

Dari grafik diatas koefisien korelasinya kecil, paling tinggi -0.3
